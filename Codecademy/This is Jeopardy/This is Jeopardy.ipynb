2. Write a function that filters the dataset for questions that contains all of the words in a list of words. For example, when the list `["King", "England"]` was passed to our function, the function returned a DataFrame of 49 rows. Every row had the strings `"King"` and `"England"` somewhere in its `" Question"`.

   Test your function by printing out the column containing the question of each row of the dataset.

````python
def return_rows(df, words):
    
    #filter must be applied to the column, it checks if all of the words in a list of words are in the string. Return True or False
    
    filter = lambda x: all(word in x for word in words)
    
    # Applies the lambda function to the Question column and returns the rows where the function returned True
    
    return df[df.question.apply(filter)]

filtered_jeo = return_rows(jeo, ["King", "England"])

print(len(filtered_jeo.question))

filtered_jeo.head()
````

3. Test your original function with a few different sets of words to try to find some ways your function breaks. Edit your function so it is more robust.

   For example, think about capitalization. We probably want to find questions that contain the word `"King"` or `"king"`.
   
   You may also want to check to make sure you don't find rows that contain substrings of your given words. For example, our function found a question that didn't contain the word `"king"`, however it did contain the word `"viking"` &mdash; it found the `"king"` inside `"viking"`. Note that this also comes with some drawbacks &mdash; you would no longer find questions that contained words like `"England's"`.

````python
# Define a function that checks if all of the words in the list are in the strings of column 'question',
# and if so return the rows
#Both capital letters and lower should be included in the solution

def filter_fun(df, words):
    filter_lambda = lambda x: all(word.lower() in x.lower() for word in words)
    return df.loc[df.question.apply(filter_lambda)]

filtered_jeo_lower = filter_fun(jeo, ["King", "England"]).reset_index()
print(len(filtered_jeo_lower.question))
filtered_jeo_lower.head()
````

````python
#exlude substrings

def filter_fun(df, words):
    filter_lambda = lambda x: all(word.lower() in x.lower().split() for word in words)
    return df.loc[df.question.apply(filter_lambda), ['show_number', 'question']]

filtered_jeo_lower_sub = filter_fun(jeo, ["King", "England"])
print(len(filtered_jeo_lower_sub.question))
filtered_jeo_lower_sub.head()
````
